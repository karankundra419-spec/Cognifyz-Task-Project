```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv("Dataset .csv")
df
```

```python
# Convert Yes/No columns into binary
df["Has Online delivery"] = (
    df["Has Online delivery"]
    .astype(str)
    .str.strip()
    .str.lower()
    .eq("yes")
    .astype(int)
)

df["Has Table booking"] = (
    df["Has Table booking"]
    .astype(str)
    .str.strip()
    .str.lower()
    .eq("yes")
    .astype(int)
)

# Ensure Price range is numeric
df["Price range"] = pd.to_numeric(df["Price range"], errors="coerce")
```

```python
task1_output = df.groupby("Price range")[[
    "Has Online delivery",
    "Has Table booking"
]].mean()

print(task1_output)
```

```python
# Online Delivery
df["Has Online delivery"] = (
    df["Has Online delivery"]
    .astype(str)
    .str.strip()
    .str.lower()
    .eq("yes")
    .astype(int)
)

# Table Booking
df["Has Table booking"] = (
    df["Has Table booking"]
    .astype(str)
    .str.strip()
    .str.lower()
    .eq("yes")
    .astype(int)
)

df["Price range"] = pd.to_numeric(df["Price range"], errors="coerce")
```

```python
relationship_analysis = df.groupby("Price range")[[
    "Has Online delivery",
    "Has Table booking"
]].mean()

print(relationship_analysis)
```

```python
relationship_analysis.plot(kind="bar", figsize=(8,5))
plt.title("Price Range vs Online Delivery & Table Booking")
plt.ylabel("Proportion")
plt.xlabel("Price Range")
plt.show()
```
